```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("test_scores.csv")
df.head()
```

```python
X = df["math"]
Y = df["cs"]
```

```python
def gradient_descent(X,Y):
    m_curr = b_curr = 0
    n = len(X)
    iterations = 1000
    learning_rate = 0.001
    for i in range(iterations):
        y_pred = m_curr*X + b_curr
#         cost = (1/n)*sum([val**2  for val in (Y-y_pred)])
        m_dev = -(2/n)*sum(X*(Y-y_pred))
        b_dev = -(1/n)*sum(Y-y_pred)
        m_curr = m_curr - learning_rate*m_dev
        b_curr= b_curr - learning_rate*b_dev
        print("m_curr: {}, b_curr: {}, iterations {}".format(m_curr,b_curr,i))
```

```python
gradient_descent(X,Y)
```
